```python
for i in range(1,2):
  url='https://www.flipkart.com/search?q={}&page={}'
  url=url.format(product_name,i)
  page=requests.get(url)
  html=page.content

  page_soup=bs4.BeautifulSoup(html,'html.parser')

  for containers in page_soup.findAll('div',{'class':'_4ddWXP'}):

    name=containers.find('a',attrs={'class':'s1Q9rs'})
    print(name)
    price=containers.find('div',attrs={'class':'_30jeq3'})
    print(price)
    rating=containers.find('div',attrs={'class':'_3LWZlK'})
    print(rating)
    packing=containers.find('div',attrs={'class':'_3Djpdu'})
    print(packing)
    #print('Specification:',specification.text)

    products.append(name.text) if type(name) == bs4.element.Tag else products.append('NaN')
    prices.append(price.text) if type(price) == bs4.element.Tag else prices.append('NaN')
    ratings.append(rating.text) if type(rating) == bs4.element.Tag  else ratings.append('NaN')
    packings.append(packing.text) if type(packing) == bs4.element.Tag else packings.append('NaN')
      # Add Rating to list

    count=count+1

  df=pd.DataFrame({'Product_name':products,'Price':prices,'Rating':ratings,'Packing':packings})
print('No of rows searched',count)
```
