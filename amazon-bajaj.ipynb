```python
import requests
import bs4
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
count=0
#container:	<div><class>sg-col-inner
#name:		<span><class>a-size-medium a-color-base a-text-normal
#rating:		<span><class>a-icon-alt
#sales:		<span><class>a-size-base
#price:		<span><class>a-price-whole
#fullprice:	<span><class>a-offscreen

names=[]
products=[]
prices=[]
ratings=[]
sales=[]
fullprices=[]

df=pd.DataFrame()
product_name='bajaj'
url='https://www.amazon.in/s?k={0}&pages=1'
print(url)
```

```python
for i in range(1,26):
  url='https://www.amazon.in/s?k={}&page={}'
  url=url.format(product_name,i)
  print(url)
```

```python
for i in range(1,4):
  
  url='https://www.amazon.in/search?q={}&page={}'
  url=url.format(product_name,i)
  print(url)
  page=requests.get(url)
  print(page)
  html=page.content

  page_soup=bs4.BeautifulSoup(html,'html.parser')
```

```python
for i in range(1,20):
  url='https://www.amazon.in/s?k={}&page={}'
  url=url.format(product_name,i)
  page=requests.get(url)
  html=page.content

  page_soup=bs4.BeautifulSoup(html,'html.parser')
  
  for containers in page_soup.findAll('div',{'id':'search'}):
  
    name=containers.find('span',attrs={'class':'a-size-medium a-color-base a-text-normal'})
    print(name)
    price=containers.find('span',attrs={'class':'a-price-whole'})
    print(price)
    rating=containers.find('span',attrs={'class':'a-icon-alt'})
    print(rating)
    sale=containers.find('span',attrs={'class':'a-size-base'})
    fullprice=containers.find('span',attrs={'class':'a-offscreen'})
    
    products.append(name.text) if type(name) == bs4.element.Tag else products.append('NaN')
    prices.append(price.text) if type(price) == bs4.element.Tag else prices.append('NaN')
    ratings.append(rating.text) if type(rating) == bs4.element.Tag  else ratings.append('NaN')
    sales.append(sale.text) if type(sale) == bs4.element.Tag else sales.append('NaN')
    fullprices.append(fullprice.text) if type(fullprice) == bs4.element.Tag else fullprices.append('NaN')
    
    count=count+1

  df=pd.DataFrame({'Product_name':products,'Price':prices,'Rating':ratings,'Sales':sales,'FullPrice':fullprices})
print('No of rows searched',count)
```
